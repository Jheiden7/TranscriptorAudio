# Filtrado y conteo de palabras.

```python
def discriminacion(song): #Se le pasa como parámetro la letra de la canción que se va a procesar.
  # Eliminar los puntos y las comas del texto:
  # Se crea un traductor utilizando el método maketrans() del tipo de dato str. Este traductor se utiliza para eliminar todos los signos 
  # de puntuación del texto. string.punctuation es una cadena predefinida en el módulo string que contiene todos los signos de puntuación.
  songTotal = song
  translator = str.maketrans('', '', string.punctuation)
  songTotal = songTotal.translate(translator)

  # Dividir el texto en palabras individuales:
  palabrasElec = songTotal.split()

  # Filtrar las palabras no deseadas:

  # Se aplica una comprensión de lista para filtrar las palabras no deseadas. Cada palabra en palabrasElec se verifica si está en palabras_filtradas 
  # Después de convertirla a minúsculas. Si la palabra no está en palabras_filtradas, se agrega a una nueva lista llamada palabras_filtradas.

  palabras_filtradas2 = [palabra for palabra in palabrasElec if palabra.lower() not in palabras_filtradas]

  # Unir las palabras filtradas en una cadena nuevamente
  songTotal_filtrada = ' '.join(palabras_filtradas2)
  #print(elecTotal_filtrada)

  return songTotal_filtrada
```

```python
def grafica(filtrado, cadena): #Se le pasa la letra de la canción ya discriminada y la cadena que complementará el título de la gráfica (p.e: "Palabras más y menos usadas *en el rap* (cadena)").
  # Detener la instancia actual de SparkContext si existe:

  elecRDD = sc.parallelize([filtrado]) # Al crear el RDD con sc.parallelize, estás indicando a Spark que divida y distribuya los elementos de suma1_filtrada en diferentes nodos para su procesamiento.

  regex = re.compile('[^a-zA-Z]') # Define una expresión regular para filtrar caracteres no alfabéticos y espacios en blanco.

  #/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////

  # Crear el RDD con  utiliza la función 'flatMap' para aplicar una operación a cada elemento del RDD (elecRDD). La función lambda toma cada línea del RDD y utiliza la expresión regular re.split('\W+', linea) 
  # para dividir la línea en palabras utilizando cualquier carácter que no sea alfanumérico (\W+) como delimitador. El resultado es un RDD plano de todas las palabras presentes en las líneas del RDD original.
  palabrasRDD = elecRDD.flatMap(lambda linea: re.split('\W+', linea)) \
                      .filter(lambda palabra: len(palabra) > 0)  # filter para eliminar las palabras vacías

  ## Contar las palabras y devolver el top 15:

  #  En la linea '26' se utiliza la función 'map' para asignar a cada palabra del RDD a palabrasRDD la forma (palabra, 1).
  palabrasContadasRDD = palabrasRDD.map(lambda palabra: (palabra, 1)) \
                                  .reduceByKey(lambda a, b: a + b)             # 'reduceByKey' para agregar la frecuencia de cada palabra sumando los valores correspondientes.
  elec15Palabrasma = palabrasContadasRDD.takeOrdered(15, key=lambda x: -x[1])  # Se almacenan las 15 palabras mas usadas.
  elec15Palabrasme = palabrasContadasRDD.takeOrdered(15, key=lambda x: x[1])   # Se almacenan las 15 palabras mas usadas.

  # Se crean listas vacías para almacenar las palabras y las frecuencias de las 15 palabras más usadas.
  palabras_mas = []
  frecuencias_mas = []
  # Se crean listas vacías para almacenar las palabras y las frecuencias de las 15 palabras menoos usadas.
  palabras_menos = []
  frecuencias_menos = []

  # Se recorre 'elec15Palabrasma' y se almacena en 'palabras_mas' y 'frecuencia_mas'.
  for palabra, frecuencia in elec15Palabrasma:
      palabras_mas.append(palabra)
      frecuencias_mas.append(frecuencia)
  # Se recorre 'elec15Palabrasme' y se almacena en 'palabras_menos' y 'frecuencia_menos'.
  for palabra, frecuencia in elec15Palabrasme:
      palabras_menos.append(palabra)
      frecuencias_menos.append(frecuencia)

  # Graficamos: 

  plt.figure(figsize=(15, 8))
  plt.bar(palabras_mas, frecuencias_mas, color='b', label='Más usadas')
  plt.bar(palabras_menos, frecuencias_menos, color='r', label='Menos usadas')
  plt.title("Palabras más y menos usadas "+cadena)
  plt.xlabel("Palabras")
  plt.ylabel("Frecuencia")
  plt.xticks(fontsize=8, rotation=45)
  plt.legend()
  plt.grid()
  plt.show()
```
